```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

df = pd.read_csv('/content/drive/MyDrive/smartphones.csv')
df.head()
```

1. Data Cleaning, jika ada nilai yang kosong atau NaN lakukan penggantian data

```python
df.info()

missing_values = df.isnull().sum()
print("\nJumlah nilai kosong atau NaN untuk setiap fitur:")
print(missing_values)
```

```python
df['avg_rating'].fillna(df['avg_rating'].mean(), inplace=True)
df['processor_speed'].fillna(df['processor_speed'].mean(), inplace=True)
df['battery_capacity'].fillna(df['battery_capacity'].mean(), inplace=True)
df['primary_camera_front'].fillna(df['primary_camera_front'].mean(), inplace=True)
df['fast_charging'].fillna(df['fast_charging'].mean(), inplace=True)

mode_os = df['os'].mode()[0]
df['os'].fillna(mode_os, inplace=True)

mode_processor_brand = df['processor_brand'].mode()[0]
df['processor_brand'].fillna(mode_processor_brand, inplace=True)

mode_num_cores = df['num_cores'].mode()[0]
df['num_cores'].fillna(mode_num_cores, inplace=True)

# DataFrame setelah data cleaning
print(df.head())

df.to_csv('smartphones_clean.csv', index=False)
```

```python
df.info()
```

2. Encoding

```python
encoded_df = df.copy()

label_encoder = LabelEncoder()

# label encoding untuk setiap kolom kategorikal
encoded_df['brand_name'] = label_encoder.fit_transform(encoded_df['brand_name'])
encoded_df['model'] = label_encoder.fit_transform(encoded_df['model'])
encoded_df['processor_brand'] = label_encoder.fit_transform(encoded_df['processor_brand'])
encoded_df['os'] = label_encoder.fit_transform(encoded_df['os'])

encoded_df
```

3. Normalisasi Data

```python
normalized_df = encoded_df.copy()

scaler = MinMaxScaler()

numeric_columns = ['price', 'avg_rating', 'processor_speed', 'battery_capacity', 'ram_capacity', 'internal_memory', 'screen_size', 'refresh_rate', 'num_rear_cameras', 'primary_camera_rear', 'primary_camera_front', 'resolution_height', 'resolution_width']
normalized_df[numeric_columns] = scaler.fit_transform(normalized_df[numeric_columns])

print("DataFrame setelah normalisasi:")
normalized_df
```

4. Data Splitting, pisahkan data menjadi features dan target. Kemudian, bagi data tersebut menjadi set pelatihan dan set pengujian dengan proporsi 70:30.

```python
# Memisahkan fitur dan target
X = normalized_df.drop(columns=['price'])  # Fitur
y = normalized_df['price']  # Target

# Memisahkan data menjadi set pelatihan dan set pengujian dengan proporsi 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Ukuran set pelatihan (X_train):", X_train.shape)
print("Ukuran set pengujian (X_test):", X_test.shape)
print("Ukuran target pelatihan (y_train):", y_train.shape)
print("Ukuran target pengujian (y_test):", y_test.shape)
```

5. Reduksi Dimensi, Terapkan PCA pada 'features' untuk mengurangi dimensi data. Pilih jumlah komponen sedemikian rupa sehingga menjelaskan setidaknya 95% varians dari data.

```python
from sklearn.decomposition import PCA
minmax_x_train = scaler.fit_transform(X_train)

# Terapkan PCA dengan mengambil komponen yang menjelaskan 95% varians data
pca = PCA(n_components=0.95)
pc = pca.fit_transform(minmax_x_train)

print('Components Variance:')
print('1st: %.3f' % pca.explained_variance_ratio_[0])
print('2nd: %.3f' % pca.explained_variance_ratio_[1])
print('3rd: %.3f' % pca.explained_variance_ratio_[2])
```
