1. Import Library yang diperlukan

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

2. Tampilkan seluruh nilai dari dataset

```python
# Tulis kode disini
df = pd.read_csv('/content/drive/MyDrive/Iklan_sosmed.csv')

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df
```

3. Lakukan EDA secukupnya

```python
# Tulis kode disini
df.info()

duplikat = df.duplicated(subset=['UserID'])

if duplikat.any():
    print("Ada duplikat data pada kolom 'UserID'.")
else:
    print("Tidak ada duplikat data pada kolom 'UserID'.")
```

analisis deskriptif

```python
df.describe()
```

data outlier

```python
Q1_age = df['Usia'].quantile(0.25)
Q3_age = df['Usia'].quantile(0.75)
IQR_age = Q3_age - Q1_age

Q1_salary = df['EstimasiGaji'].quantile(0.25)
Q3_salary = df['EstimasiGaji'].quantile(0.75)
IQR_salary = Q3_salary - Q1_salary

# batas outlier
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age

lower_bound_salary = Q1_salary - 1.5 * IQR_salary
upper_bound_salary = Q3_salary + 1.5 * IQR_salary

# Filter data outlier
outliers_age = df[(df['Usia'] < lower_bound_age) | (df['Usia'] > upper_bound_age)]
outliers_salary = df[(df['EstimasiGaji'] < lower_bound_salary) | (df['EstimasiGaji'] > upper_bound_salary)]

# Tampilkan data outlier
print("Outlier pada fitur Usia:")
print(outliers_age)

print("\nOutlier pada fitur EstimasiGaji:")
print(outliers_salary)

# Boxplot untuk fitur Usia
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Usia'])
plt.title('Boxplot Usia')
plt.show()

# Boxplot untuk fitur EstimasiGaji
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['EstimasiGaji'])
plt.title('Boxplot Estimasi Gaji')
plt.show()
```

4. Cek apakah ada nilai yang hilang pada dataset dan isi nilai yang hilang pada 'EstimasiGaji' dengan median dari kolom tersebut.

```python
# Tulis kode disini
missing_values = df.isnull().sum()
print("Jumlah nilai yang hilang pada setiap kolom:")
print(missing_values)

df['EstimasiGaji'].fillna(int(df['EstimasiGaji'].median()), inplace=True)
```

5. Lakukan encoding pada kolom 'Kelamin'

```python
# Tulis kode disini
label_encoder = LabelEncoder()

df['Kelamin_encoded'] = label_encoder.fit_transform(df['Kelamin'])
df['Kelamin'] = df['Kelamin_encoded']
df.drop('Kelamin_encoded', axis=1, inplace=True)
df.head(10)
```

6. Standarisasi kolom 'Usia' dan 'EstimasiGaji' menggunakan StandardScaler.

```python
# Tulis kode disini
scaler = StandardScaler()

df[['Usia', 'EstimasiGaji']] = scaler.fit_transform(df[['Usia', 'EstimasiGaji']])
#print(df[['Usia', 'EstimasiGaji']])
df.head(10)
```

7. Urutkan dataset berdasarkan 'EstimasiGaji' dari yang terendah ke tertinggi.

```python
# Tulis kode disini
df_sorted = df.sort_values(by='EstimasiGaji')

df_sorted
```

8. Simpan dataset yang telah dibersihkan dan diproses ke file baru bernama 'Iklan_sosmed_clean.csv'

```python
# Tulis kode disini
file_path = '/content/drive/My Drive/Iklan_sosmed_clean1.csv'

df_sorted.to_csv(file_path, index=False)
print("Dataset telah disimpan di Google Drive dengan nama:", file_path)
```
