2. Menganalisis business problem/question (minimal 5)

1. Bagaimana distribusi harga smartphone di dalam dataset?


```python
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribusi Harga Smartphone (Histogram)')
plt.xlabel('Harga')
plt.ylabel('Frekuensi')
plt.show()
```

2. Apakah merek tertentu mendominasi pasar smartphone?


```python
brand_counts = df['brand_name'].value_counts()

plt.figure(figsize=(10, 6))
brand_counts.plot(kind='bar')
plt.title('Distribusi Merek Smartphone dalam Dataset')
plt.xlabel('Merek Smartphone')
plt.ylabel('Frekuensi')
plt.xticks(rotation=45, ha='right')
plt.show()
```

3. Bagaimana perbandingan rata-rata harga smartphone antara merek-merek yang ada di pasaran?

```python
average_price_by_brand = df.groupby('brand_name')['price'].mean()
print(average_price_by_brand)

plt.figure(figsize=(12, 6))
sns.barplot(x=average_price_by_brand.index, y=average_price_by_brand.values)
plt.title('Perbandingan Rata-rata Harga Smartphone Berdasarkan Merek')
plt.xlabel('Merek Smartphone')
plt.ylabel('Rata-rata Harga')
plt.xticks(rotation=60)
plt.show()
```

4. Apakah terdapat korelasi antara rating pengguna dan harga smartphone?


```python
df_clean = df.dropna(subset=['battery_capacity', 'price'])

plt.figure(figsize=(10, 6))
sns.scatterplot(x='battery_capacity', y='price', data=df_clean)
plt.title('Hubungan Antara Kapasitas Baterai dan Harga Smartphone')
plt.xlabel('Kapasitas Baterai (mAh)')
plt.ylabel('Harga')
plt.show()

correlation = df_clean['battery_capacity'].corr(df_clean['price'])
print("Korelasi antara kapasitas baterai dan harga smartphone:", correlation)
```

5. Bagaimana distribusi ukuran penyimpanan (storage) pada smartphone?


```python
storage_counts = df['internal_memory'].value_counts()
print("Internal Memory")
print(storage_counts)

plt.figure(figsize=(10, 6))
plt.hist(df['internal_memory'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribusi Ukuran Penyimpanan Smartphone')
plt.xlabel('Ukuran Penyimpanan (GB)')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```

6. Apakah ada hubungan antara rating kamera dan rating umum smartphone?

```python
df_clean = df.dropna(subset=['avg_rating', 'primary_camera_rear'])

plt.figure(figsize=(10, 6))
sns.scatterplot(x='primary_camera_rear', y='avg_rating', data=df_clean)
plt.title('Hubungan Antara Rating Kamera dan Rating Umum Smartphone')
plt.xlabel('Rating Kamera')
plt.ylabel('Rating Umum')
plt.show()

correlation = df_clean['primary_camera_rear'].corr(df_clean['avg_rating'])
print("Korelasi antara rating kamera dan rating umum smartphone:", correlation)
```

3. Menampilkan analisis deskriptif statistik / data outlier

```python
df.describe()
```

data outlier

```python
import seaborn as sns
import matplotlib.pyplot as plt

#menggunakan Interquartile Range (IQR)
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
print("Data Outlier pada kolom 'price':")
print(outliers)

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'])
plt.title('Box Plot Harga Smartphone')
plt.xlabel('Harga')
plt.show()
```
